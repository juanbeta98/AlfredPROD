```python
import json

# Basic way to load a JSON file
with open('./input.json', 'r') as f:
    data = json.load(f)

```

```python
for i in data['results']:
    if i['state'] not in  ['IN_PROGRESS', 'COMPLETED']:
        print(f'{i["state"]} - {i["scheduleDate"]}')
        for j in i['serviceLabors']:
            print(f'- {j["startDate"]} - {j["endDate"]}')
```
